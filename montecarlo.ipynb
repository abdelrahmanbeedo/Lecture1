```python
import numpy as np
import matplotlib.pyplot as plt

init_investment = 20000
mean_return = 0.04
std_deviation = 0.13
years = 6
simulations = 7500

protofolio_values = np.zeros((simulations, years))


for i in range(simulations):
  annual_returns = np.random.normal(mean_return, std_deviation, years)
  protofolio_values[i] = init_investment * np.cumprod(1 + annual_returns)

  mean_final_value = np.mean(protofolio_values[:, -1])
  std_final_value = np.std(protofolio_values[:, -1])
  median_final_value = np.median(protofolio_values[:, -1])


# Plot histogram of final portfolio values
plt.figure(figsize=(10, 6))
plt.hist(protofolio_values[:, -1], bins=50, color='skyblue', edgecolor='black')
plt.axvline(mean_final_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: ${mean_final_value:,.2f}')
plt.axvline(median_final_value, color='green', linestyle='dashed', linewidth=2, label=f'Median: ${median_final_value:,.2f}')
plt.title('Distribution of Final Portfolio Values')
plt.xlabel('Final Portfolio Value ($)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```
